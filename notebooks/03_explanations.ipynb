```python
import pandas as pd
import numpy as np
import joblib
import shap
import os
from lime.lime_tabular import LimeTabularExplainer
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler

# === CONFIGURATION ===
ratios = ["1", "2", "5", "10", "15", "20", "30", "40", "45", "50"]
n_iterations = 100
random_seed = 42
target_set_path = "data/target_set.csv"
sample_dir = "data/sampled"
model_dir = "results"
results_dir = "results/iterations"
os.makedirs(results_dir, exist_ok=True)

# === LOAD TARGET SET ===
target_df = pd.read_csv(target_set_path)
X_target = target_df.drop(columns=["default"]).values

# === MAIN LOOP ===
for ratio in ratios:
    print(f"🔄 Running explanations for ratio {ratio}%...")

    # Load training set and fit LIME explainer
    train_df = pd.read_csv(f"{sample_dir}/train_{ratio}.csv")
    X_train = train_df.drop(columns=["default"]).values
    y_train = train_df["default"].values

    lime_explainer = LimeTabularExplainer(
        training_data=X_train,
        feature_names=train_df.drop(columns="default").columns.tolist(),
        class_names=["non-default", "default"],
        discretize_continuous=True,
        mode="classification"
    )

    # Load trained model
    model_path = f"{model_dir}/xgb_model_{ratio}.pkl"
    model = joblib.load(model_path)

    # Prepare to collect explanation results
    all_shap_vals = []
    all_lime_vals = []

    for i in range(n_iterations):
        print(f"    ➤ Iteration {i+1}/{n_iterations}", end="\r")

        # SHAP Explanation
        shap_explainer = shap.Explainer(model.predict_proba, X_train, feature_names=train_df.drop(columns="default").columns.tolist())
        shap_values = shap_explainer(X_target)
        shap_instance_vals = [np.abs(shap_values[j].values[:, 1]) for j in range(len(shap_values))]
        all_shap_vals.append(pd.DataFrame(shap_instance_vals).assign(iteration=i))

        # LIME Explanation
        lime_instance_vals = []
        for instance in X_target:
            exp = lime_explainer.explain_instance(instance, model.predict_proba, num_features=X_train.shape[1])
            instance_exp = dict(exp.as_list())
            lime_vals = [np.abs(instance_exp.get(f, 0)) for f in lime_explainer.feature_names]
            lime_instance_vals.append(lime_vals)

        all_lime_vals.append(pd.DataFrame(lime_instance_vals).assign(iteration=i))

    # Save outputs
    pd.concat(all_shap_vals).to_csv(f"{results_dir}/shap_iterations_{ratio}.csv", index=False)
    pd.concat(all_lime_vals).to_csv(f"{results_dir}/lime_iterations_{ratio}.csv", index=False)

    print(f"✅ Saved SHAP and LIME iteration results for {ratio}% at: {results_dir}/")
```
