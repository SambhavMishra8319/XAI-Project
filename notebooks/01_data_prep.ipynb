```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import os

# Load dataset
df = pd.read_excel("../data/default_of_credit_card_clients.xls", header=1)
df.rename(columns={"default payment next month": "default"}, inplace=True)

# Drop ID column if exists
if "ID" in df.columns:
    df.drop(columns=["ID"], inplace=True)

# Scale numeric features
scaler = StandardScaler()
X = df.drop(columns=["default"])
y = df["default"]
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Combine for convenience
df_scaled = X_scaled.copy()
df_scaled["default"] = y.values

# Fix random seed for reproducibility
RANDOM_STATE = 42

# Sample 100 defaults and 100 non-defaults for explanation target set
target_defaults = df_scaled[df_scaled["default"] == 1].sample(n=100, random_state=RANDOM_STATE)
target_nondefaults = df_scaled[df_scaled["default"] == 0].sample(n=100, random_state=RANDOM_STATE)
target_set = pd.concat([target_defaults, target_nondefaults]).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

# Save target set and training pool
os.makedirs("data", exist_ok=True)
target_set.to_csv("data/target_set.csv", index=False)
df_scaled.drop(target_set.index).to_csv("data/train_pool.csv", index=False)

# Create result folders
os.makedirs("results", exist_ok=True)
os.makedirs("plots", exist_ok=True)

print("✅ Target set and training pool saved.")
```
