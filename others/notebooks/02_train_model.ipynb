```python
import pandas as pd
import xgboost as xgb
import joblib
import os
from sklearn.metrics import accuracy_score

# Paths
sample_dir = "./data/sampled"
results_dir = "./results"
os.makedirs(results_dir, exist_ok=True)

# Load target set to evaluate model (optional sanity check)
target_set = pd.read_csv("./data/target_set.csv")
X_target = target_set.drop(columns=["default"])
y_target = target_set["default"]

# Train on each imbalance setting
ratios = ["1", "2", "5", "10", "15", "20", "30", "40", "45", "50"]

for ratio in ratios:
    train_file = os.path.join(sample_dir, f"train_{ratio}.csv")
    if not os.path.exists(train_file):
        print(f"⚠️ Missing {train_file}. Skipping.")
        continue

    # Load and split
    df = pd.read_csv(train_file)
    X = df.drop(columns=["default"])
    y = df["default"]

    # Train model
    model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
    model.fit(X, y)

    # Evaluate on fixed target set (optional)
    preds = model.predict(X_target)
    acc = accuracy_score(y_target, preds)
    print(f"✅ Trained XGB model at {ratio}% default rate – Accuracy: {acc:.3f}")

    # Save model
    joblib.dump(model, os.path.join(results_dir, f"xgb_model_{ratio}.pkl"))
```
