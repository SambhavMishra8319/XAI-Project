```python
import pandas as pd
import numpy as np
from sklearn.utils import resample
import os

train_pool = pd.read_csv("data/train_pool.csv")

# Configuration
RANDOM_STATE = 42
ratios = [0.01, 0.025, 0.05, 0.10, 0.15, 0.20, 0.30, 0.40, 0.45, 0.50]
Z = 6258  # fixed sample size from paper
save_path = "data/sampled"

os.makedirs(save_path, exist_ok=True)

for ratio in ratios:
    pos_count = int(Z * ratio)
    neg_count = Z - pos_count

    pos_samples = train_pool[train_pool["default"] == 1]
    neg_samples = train_pool[train_pool["default"] == 0]

    if len(pos_samples) < pos_count or len(neg_samples) < neg_count:
        print(f"⚠️ Not enough samples for ratio {ratio*100:.1f}%. Skipping.")
        continue

    sampled_pos = pos_samples.sample(n=pos_count, random_state=RANDOM_STATE)
    sampled_neg = neg_samples.sample(n=neg_count, random_state=RANDOM_STATE)
    sampled_df = pd.concat([sampled_pos, sampled_neg]).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

    sampled_df.to_csv(f"{save_path}/train_{int(ratio*100)}.csv", index=False)
    print(f"✅ Saved train_{int(ratio*100)}.csv with default rate {ratio*100:.1f}%")
```
